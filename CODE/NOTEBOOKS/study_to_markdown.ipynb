````python
FILE_OUTPUT_MD = os.path.join(DIR_REF_STUDY, STUDY_REF + '.md')
FILE_OUTPUT_HTML = os.path.join(DIR_REF_STUDY, STUDY_REF + '.html')
#FILE_OUTPUT_PDF = os.path.join(DIR_REF_STUDY, STUDY_REF + '.pdf')
````

````python
FILE_TMP_DOWNLOAD = os.path.join(DIR_REF_STUDY, "download_" + STUDY_REF + ".zip")
````

````python
def add_image(root_folder, filename_img):
  path = os.path.join(root_folder, filename_img)
  res = "![Alt text]({})".format(path) + '\n'
  return res

def add_code(language, code):
  res = '```'
  res += language + '\n'
  res += code +'\n'
  res += '```' + '\n'
  return res
````

````python
add_image('.', 'test.svg')
````

````python
add_code('json', "{a: 10, b:20}")
````

````python
def add_trial(name_trial, path):
  file_local = open(os.path.join(path, 'local_setting.json'), 'r')

  res = "### " + name_trial + '\n'
  res += add_code('json', file_local.read())

  res += add_image('./TRIALS/'+name_trial, 'plot.svg')
  res += add_image('./TRIALS/'+name_trial, 'plot_acc_f1.svg')

  file_local.close()
  return res
````

````python
file_out = open(FILE_OUTPUT_MD, 'w')
file_global_sett = open(os.path.join(DIR_REF_STUDY, 'global_setting.json'), 'r')
file_best_info = open(os.path.join(DIR_REF_STUDY, 'best_trial_info.json'), 'r')


file_out.write("# " + STUDY_REF + '\n')
file_out.write(add_code('json', file_global_sett.read()))
file_out.write("\n\n")

file_out.write("## BEST PARAMETERS:\n")
file_out.write(add_code('json', file_best_info.read()))
file_out.write("\n\n")

for filename in os.listdir(DIR_TRIALS):
  file_out.write('\n**********\n')
  file_out.write(add_trial(filename, os.path.join(DIR_TRIALS, filename)))
  file_out.write('\n**********\n')
file_out.close()
````
